```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("trainMetrics.csv")
```

```python
df.plot(figsize=(20,10), kind = 'scatter',x='iteration',y='total_loss') # scatter plot
```

```python
data_path = "trainMetrics.csv"
data = np.genfromtxt(data_path, delimiter=',', names = True)
```

```python
%matplotlib inline

def func(x,a,b,c):
    return a * np.exp(-b * x) + c

def func2(x, a, b):
    return a * np.exp(-b * x)

def func3(x, a, b):
    return a * b ** x
```

```python
def func1(x, a, b):
    return a * b ** x
```

```python
fig= plt.figure(figsize=(60,30))

popt, pcov = curve_fit(func3, data['iteration'], data['total_loss'], maxfev = 1000 )
x_plot=np.linspace(0,179999,100)
plt.plot(x_plot, func3(x_plot, *popt), 'r-')
plt.scatter(data['iteration'],data['total_loss'],label='iterations vs loss')

plt.title("a*b^x")

plt.xlabel('iteration')
plt.ylabel('loss')
plt.legend()
plt.show()

```

```python
fig= plt.figure(figsize=(60,30))

popt, pcov = curve_fit(func2, data['iteration'], data['total_loss'], p0=[1, 1e-6] )
x_plot=np.linspace(0,179999,100)
plt.plot(x_plot, func2(x_plot, *popt), 'r-')
plt.scatter(data['iteration'],data['total_loss'],label='iterations vs loss')

plt.title("a*b^x")

plt.xlabel('iteration')
plt.ylabel('loss')
plt.legend()
plt.savefig('plot')
plt.show()
```
